```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o arquivo clientes.csv
clientes = pd.read_csv('clientes.csv', parse_dates=['data_inclusao'])

# Extraindo o ano de inclusão dos clientes
clientes['ano_inclusao'] = clientes['data_inclusao'].dt.year

# Agrupando os clientes por ano de inclusão e contando a quantidade em cada grupo
clientes_por_ano = clientes.groupby('ano_inclusao').size()

# Filtrando os dados para anos a partir de 2010
clientes_por_ano = clientes_por_ano[clientes_por_ano.index >= 2010]

# Criando um gráfico de linhas com a evolução do número de clientes
plt.figure(figsize=(10, 5))
plt.plot(clientes_por_ano.index, clientes_por_ano.values)
plt.xlabel('Ano')
plt.ylabel('Quantidade de Clientes')
plt.title('Evolução do Número de Clientes (a partir de 2010)')
plt.xticks(clientes_por_ano.index)  # Ajustando os ticks do eixo x para os anos
#plt.grid(True)  # Adicionando grade ao gráfico

# Adicionando a quantidade de clientes em cada ponto do gráfico
for i, v in enumerate(clientes_por_ano.values):
    plt.text(clientes_por_ano.index[i], v, str(v), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o arquivo clientes.csv
clientes = pd.read_csv('clientes.csv', parse_dates=['data_inclusao'])

# Extraindo o ano de inclusão dos clientes
clientes['ano_inclusao'] = clientes['data_inclusao'].dt.year

# Agrupando os clientes por ano de inclusão e contando a quantidade em cada grupo
clientes_por_ano = clientes.groupby('ano_inclusao').size()

# Calculando o número total acumulado de clientes por ano
clientes_por_ano_acumulado = clientes_por_ano.cumsum()

# Criando um gráfico de barras com a quantidade de clientes por ano
plt.figure(figsize=(10, 5))
plt.bar(clientes_por_ano_acumulado.index, clientes_por_ano_acumulado.values)
plt.xlabel('Ano')
plt.ylabel('Quantidade de Clientes')
plt.title('Evolução do Número Total de Clientes por Ano')
plt.xticks(clientes_por_ano_acumulado.index)  # Ajustando os ticks do eixo x para os anos
#plt.grid(True)  # Adicionando grade ao gráfico

# Adicionando a quantidade de clientes em cada barra
for i, v in enumerate(clientes_por_ano_acumulado.values):
    plt.text(clientes_por_ano_acumulado.index[i], v, str(v), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o arquivo clientes.csv
clientes = pd.read_csv('clientes.csv', parse_dates=['data_nascimento'])

# Calculando a idade dos clientes
clientes['idade'] = (pd.Timestamp('today') - clientes['data_nascimento']).dt.days / 365.25
clientes['idade'] = clientes['idade'].astype(int)

# Definindo faixas etárias
faixas_etarias = pd.cut(clientes['idade'], bins=[0, 18, 30, 50, 70, 150], labels=['0-18', '18-30', '30-50', '50-70', '70+'])

# Agrupando os clientes por faixa etária e contando a quantidade em cada grupo
clientes_por_faixa_etaria = clientes.groupby(faixas_etarias).size()

# Criando um gráfico de barras com a quantidade de clientes por faixa etária
plt.figure(figsize=(10, 5))
plt.bar(clientes_por_faixa_etaria.index, clientes_por_faixa_etaria.values)
plt.xlabel('Faixa Etária')
plt.ylabel('Quantidade de Clientes')
plt.title('Quantidade de Clientes por Faixa Etária')
plt.xticks(rotation=45)  # Rotacionando os labels do eixo x para melhor visualização

# Adicionando a quantidade de clientes em cada barra
for i, v in enumerate(clientes_por_faixa_etaria.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```
