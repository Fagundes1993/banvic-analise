```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o arquivo colaborador_agencia.csv
colaboradores_agencia = pd.read_csv('colaborador_agencia.csv')

# Dicionário com os nomes das agências
nomes_agencias = {
    1: 'Agência Matriz',
    2: 'Agência Tatuapé',
    3: 'Agência Campinas',
    4: 'Agência Osasco',
    5: 'Agência Porto Alegre',
    6: 'Agência Rio de Janeiro',
    7: 'Agência Digital',
    8: 'Agência Jardins',
    9: 'Agência Florianópolis',
    10: 'Agência Recife'
}

# Criando uma coluna 'nome_agencia' com base no dicionário
colaboradores_agencia['nome_agencia'] = colaboradores_agencia['cod_agencia'].map(nomes_agencias)

# Agrupando os colaboradores por 'nome_agencia' e contando a quantidade em cada grupo
colaboradores_por_agencia = colaboradores_agencia.groupby('nome_agencia').size()

# Criando um gráfico de barras com a quantidade de colaboradores por agência
plt.figure(figsize=(12, 6))
plt.bar(colaboradores_por_agencia.index, colaboradores_por_agencia.values)
plt.xlabel('Agência')
plt.ylabel('Quantidade de Colaboradores')
plt.title('Distribuição de Colaboradores por Agência')
plt.xticks(rotation=45)  # Rotacionando os labels do eixo x para melhor visualização

# Adicionando a quantidade de colaboradores em cada barra
for i, v in enumerate(colaboradores_por_agencia.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o arquivo colaborador_agencia.csv
colaborador_agencia = pd.read_csv('colaborador_agencia.csv')

# Dicionário com os nomes das agências
nomes_agencias = {
    1: 'Matriz',
    2: 'Tatuapé',
    3: 'Campinas',
    4: 'Osasco',
    5: 'Porto Alegre',
    6: 'Rio de Janeiro',
    7: 'Digital',
    8: 'Jardins',
    9: 'Florianópolis',
    10: 'Recife'
}

# Criando uma coluna 'nome_agencia' com base no dicionário
colaborador_agencia['nome_agencia'] = colaborador_agencia['cod_agencia'].map(nomes_agencias)

# Agrupando os colaboradores por 'nome_agencia' e contando a quantidade em cada grupo
colaboradores_por_agencia = colaborador_agencia.groupby('nome_agencia').size()

# Ordenando as agências em ordem crescente de quantidade de colaboradores
colaboradores_por_agencia = colaboradores_por_agencia.sort_values()

# Criando um gráfico de barras com a quantidade de colaboradores por agência
plt.figure(figsize=(10, 5))
plt.bar(colaboradores_por_agencia.index, colaboradores_por_agencia.values)
plt.xlabel('Agência')
plt.ylabel('Quantidade de Colaboradores')
plt.title('Distribuição de Colaboradores por Agência')
plt.xticks(rotation=45)  # Rotacionando os labels do eixo x para melhor visualização

# Adicionando a quantidade de colaboradores em cada barra
for i, v in enumerate(colaboradores_por_agencia.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```
